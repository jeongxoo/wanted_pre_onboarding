```python
import os
import test_main as tm
import numpy as np
```

```python
f = open(os.path.join("test_data.txt"), "r")
text = f.read()
text = text.split(".")
text = [t.replace("\n", "").strip() for t in text]
text = text[:-2]
f.close()
```

```python
tk = tm.Tokenizer()
tfidf = tm.TfidfVectorizer(tk)
```

```python
tokenized_text = tk.preprocessing(text)
```

```python
tfidf_value = tfidf.fit_transform(text)
```

```python
for i,v in enumerate(tokenized_text):
    for j,t in enumerate(tfidf_value[i]):
        print(tokenized_text[i][j], t, end = " | ")
    print()
    max_val = max(tfidf_value[i])
    print(max_val)
    where = np.where(tfidf_value[i] == max_val)
    for w in where:
        for _w in w:
            print(v[_w])

    print()
    print()
```
